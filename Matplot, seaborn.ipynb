```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cars = pd.read_excel("C:\\Users\\Bruger\\Desktop\\car.xlsx")
```

```python
print(cars)
```

```python
sorted_data = cars.sort_values(by=['Cost'])

plt.figure(figsize=(20, 6))  

plt.plot(sorted_data['Cost'], sorted_data['Fuel'], marker='o')
plt.xlabel('Вартість палива, $')
plt.ylabel('Споживання палива, mpg')
plt.title('Залежність споживання палива від вартості')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(20, 6))  

plt.scatter(cars['Speed'], cars['Fuel'], color='green')
plt.xlabel('Швидкість, mph')
plt.ylabel('Споживання палива, mpg')
plt.title('Залежність споживання палива від швидкості')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(20, 6))  

plt.scatter(cars['Speed'], cars['CC'], color='red')
plt.xlabel('Швидкість, mph')
plt.ylabel('Вантажнопідйомність, kg')
plt.title('Залежність вантажнопідйомності від швидкості')
plt.grid(True)
plt.show()
```

```python
num_bins = 5
speed_bins = pd.cut(cars['Speed'], bins=num_bins)

plt.figure(figsize=(20, 6))  

plt.hist(cars['Speed'], bins=num_bins, edgecolor='black')
plt.xlabel('Швидкість')
plt.ylabel('Кількість автомобілів')
plt.title('Розподіл автомобілів за швидкістю')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(20, 6))  

sns.boxplot(x='Brand', y='Speed', data=cars)
plt.xlabel('Марка автомобіля')
plt.ylabel('Швидкість')
plt.title('Розподіл швидкостей за марками автомобілів')
plt.grid(True)
plt.show()
```
