```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cars = pd.read_excel("C:\\Users\\Bruger\\Desktop\\car.xlsx")
```

```python
print(cars)
```

```python
sorted_data = cars.sort_values(by=['Cost'])

plt.figure(figsize=(20, 6))  

plt.plot(sorted_data['Cost'], sorted_data['Fuel'], marker='o')
plt.xlabel(' , $')
plt.ylabel(' , mpg')
plt.title('    ')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(20, 6))  

plt.scatter(cars['Speed'], cars['Fuel'], color='green')
plt.xlabel(', mph')
plt.ylabel(' , mpg')
plt.title('    ')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(20, 6))  

plt.scatter(cars['Speed'], cars['CC'], color='red')
plt.xlabel(', mph')
plt.ylabel(', kg')
plt.title('   ')
plt.grid(True)
plt.show()
```

```python
num_bins = 5
speed_bins = pd.cut(cars['Speed'], bins=num_bins)

plt.figure(figsize=(20, 6))  

plt.hist(cars['Speed'], bins=num_bins, edgecolor='black')
plt.xlabel('')
plt.ylabel(' ')
plt.title('   ')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(20, 6))  

sns.boxplot(x='Brand', y='Speed', data=cars)
plt.xlabel(' ')
plt.ylabel('')
plt.title('    ')
plt.grid(True)
plt.show()
```
